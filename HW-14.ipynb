以下為fs.conf文件內容

device /dev/video0
resolution 640x360
title "AIOT-HW14"
jpeg 50
save ./webcam/webcam%H-%M-%S.jpg

```python
# hw14.py
import os
import time

while True:
    os.system("fswebcam -c fs.conf")
    time.sleep(15)
```
