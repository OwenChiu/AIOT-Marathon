作業1 : 練習使用一個民生物聯網 API，例如空氣、地震等感測站所回傳的紀錄資料。
參考資料網址： https://ci.taiwan.gov.tw/dsp/environmental.aspx

```python
import requests

URL = "https://sta.ci.taiwan.gov.tw/STA_Earthquake_v2/v1.0/"
response = requests.get(URL)
result = response.json()

print(result)
```
